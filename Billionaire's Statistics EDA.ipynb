## Top 5 Business Categories by Net Worth

<p>The bar plot below illustrates the top 5 business categories by their total net worth in billions. Each bar represents a business category, with its length corresponding to the aggregated net worth of all individuals within that category.</p>


```python
# Convert Final Worth in Billion
data['final_worth_in_B'] = (data['finalWorth']/1000)

# Find the top 5 Category
category_net_worth = data.groupby('category')['final_worth_in_B'].sum()
top_catagories = category_net_worth.sort_values(ascending = False).head(5)

# Extract top 5 Category name and Total Worth
catergories = top_catagories.index
total_worth = top_catagories.values

#Create a Bar Plot
plt.figure(figsize = (5,3))
plt.barh(catergories, total_worth, color = 'royalblue')
plt.ylabel('Business Category')
plt.xlabel('Total Worth (in billions)')
plt.title('Top 5 Business Categories by Net Worth')
plt.xticks(rotation=45)
plt.show()
```

## Top 5 Countries by Net Worth

<p>The pie chart below illustrates the top 5 countries by their total net worth in billions. Each slice of the pie represents a country, with its size corresponding to the aggregated net worth of all individuals within that country.</p>


```python
# Convert Final Worth in Billion
data['final_worth_in_B'] = (data['finalWorth']/1000)

# Find the top 5 Countries
country_group = data.groupby('country')['final_worth_in_B'].sum()
top_countries = country_group.sort_values(ascending = False).head(5)

# Extract top 5 countries name and Total Worth
countries = top_countries.index
total_worth = top_countries.values

#Create a Pie Plot
plt.figure(figsize = (4,4))
plt.pie(total_worth, labels = countries, autopct='%1.1f%%', startangle=140)
plt.show()
```

## Top 5 Youngest Billionaires with Their Ranks

<p>The horizontal bar chart below displays the top 5 youngest billionaires along with their ranks. Each bar represents a billionaire, with its length corresponding to the individual's age.</p>

```python
%matplotlib notebook
# Find the 5 youngest billionaires
youngest_billionaires = data.nsmallest(5, 'age')

# Extract the names, ages, and ranks of the 5 youngest billionaires
names = youngest_billionaires['personName']
ages = youngest_billionaires['age']
ranks = youngest_billionaires['rank']
worth = youngest_billionaires['finalWorth']

# Create the bar chart
fig, ax = plt.subplots(figsize=(5, 3))
bars = ax.barh(names, ages, color='royalblue')
ax.set_xlabel('Age')
ax.set_ylabel('Name')
ax.set_title('Top 5 Youngest Billionaires with Their Ranks')

# Add hover functionality using mplcursors
cursor = mplcursors.cursor(bars, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(f'Rank: {ranks.iloc[sel.target.index]}\nNet Worth: ${worth.iloc[sel.target.index]:.1f}B'))

# Display the bar chart
plt.tight_layout()
plt.show()
```

## Distribution of Self-Made vs Inherited Billionaires

<p>The pie chart below illustrates the distribution of self-made versus inherited billionaires in the dataset. Each slice of the pie represents the proportion of billionaires who are self-made compared to those who inherited their wealth.</p>


```python
# Calculate the counts of each value in the 'SelfMade' column
self_made_counts = data['selfMade'].value_counts()

# Create the pie chart
plt.figure(figsize=(4, 4))
self_made_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['LightCoral','MediumSeagreen'], labels=['Self-Made', 'Inherited'], legend=False)

# Add title
plt.title('Distribution of Self-Made vs Inherited Billionaires')

plt.ylabel('')
plt.show()
```

# Distribution of Self-Made vs Inherited Billionaires by Gender

<p>The set of pie charts below illustrates the distribution of self-made versus inherited billionaires, categorized by gender. Each pie chart represents the proportion of self-made and inherited billionaires within a specific gender category.</p>


```python
gender_counts = data['gender'].value_counts()

# Male distribution
male_self_made_counts = data[data['gender'] == 'M']['selfMade'].value_counts()

# Female distribution
female_self_made_counts = data[data['gender'] == 'F']['selfMade'].value_counts()

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(9, 3))

# Overall pie chart
axs[0].pie(gender_counts, autopct='%1.1f%%', startangle=90, colors=['LightCoral','MediumSeagreen'], labels=['Male', 'Female'])
axs[0].set_title('Gender Distribution')

# Male pie chart
axs[1].pie(male_self_made_counts, autopct='%1.1f%%', startangle=90, colors=['LightCoral','MediumSeagreen'], labels=['Self-Made', 'Inherited'])
axs[1].set_title('Self Made Male')

# Female pie chart
axs[2].pie(female_self_made_counts, autopct='%1.1f%%', startangle=90, colors=['LightCoral','MediumSeagreen'], labels=['Self-Made', 'Inherited'])
axs[2].set_title('Self Made Female')

# Adjust layout
plt.tight_layout()
plt.show()
```
